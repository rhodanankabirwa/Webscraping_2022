```python
# Scrapeing the Yelp webpage

driver = webdriver.Chrome() 
URLString = "https://www.yelp.com/search?find_desc=boston+restaurants&find_loc=Boston%2C+MA+02122&start=$NUM$"
URL = URLString.replace("$NUM$",str(0))
print("Page 1:" + URL)
driver.get(URL)  # Gets the page
page_content = driver.page_source 
```

```python
# Initialize Database

conn = sqlite3.connect('YelpV2.db')
c = conn.cursor()
c.execute("CREATE TABLE Yelp(ID INTEGER PRIMARY KEY AUTOINCREMENT, Name TEXT, StarRating REAL, NumberOfReview INTEGER, Categories TEXT, PriceRange TEXT, Neighborhood TEXT, ClosingTime TEXT, Review TEXT)")
```

```python
# RegEx

# Page Information

RegexPage = '<div class=" border-color--default__09f24__NPAKY text-align--center__09f24__fYBGO"><span class=" css-chan6m">([0-9]+) of (24)<\/span><\/div>'
PageMatch = re.compile(RegexPage).findall(page_content)
print(PageMatch)

Page = int(PageMatch[0][0].strip())
LastPage = int(PageMatch[0][1].strip())
print("There are", LastPage, "total pages that will be scraped")

 
# Incrementor
i = 0
```

```python
# Looping 23x (current + the remaining)

for pagenumber in range(LastPage):
    
    # Empty list for processing
    test = []
    
    Regex0 = '<div class=" border-color--default__09f24__NPAKY"><h3 class="css-1agk4wl"><span class=" css-1egxyvc" data-font-weight="bold">([0-9]+)<!-- -->.&nbsp;<a href="\/biz\/.*?\?osq=boston\+restaurants" class="css-1m051bw" target="_blank" name=".*?" rel="noopener">.*?<\/a><\/span><\/h3><\/div>'
    match0 = re.compile(Regex0).findall(page_content)
    
    
    
    for x in range(len(match0)):
        IdValue = [match0[x]]
        test.append(IdValue)
        
    
    
    # RegEx for the Restaurant Names
    Regex1 = '<a href="\/biz\/.*?class="css-1m051bw" target="_blank" name=".*?" rel="noopener">(.*?)<\/a>'
    match1 = re.compile(Regex1).findall(page_content)
    
    
    
    for x in range(len(match1[0:10])):
        NameValue = [match1[x]]
        test[x].extend(NameValue)
        
    

    # RegEx for the Star Rating
    Regex2 = '<div class=" five-stars.*?role="img" aria-label="(.*?) star rating">'
    match2 = re.compile(Regex2).findall(page_content)
    
    
    
    for x in range(len(match2[4:14])):
        StarValue = [match2[x + 4]]
        test[x].extend(StarValue)
        
    

    # RegEx for the Number of Reviews
    Regex3 = '<div class=" display--inline-block__09f24__fEDiJ border-color--default__09f24__NPAKY"><span class=" css-gutk1c" data-font-weight="semibold"><\/span><span class=" css-chan6m">([0-9]+)<\/span><\/div>'
    match3 = re.compile(Regex3).findall(page_content)
    
    

    for x in range(len(match3[2:12])):
        NumReviewValue = [match3[x + 2]]
        test[x].extend(NumReviewValue)


    # RegEx for Categories
    Regex4 = '<span class="css-11bijt4">(.*?)<\/span>'
    match4 = re.compile(Regex4).findall(page_content)
    
    

    for x in range(len(match4[0:10])):
        CategoryReviewValue = [match4[x]]
        test[x].extend(CategoryReviewValue)
    

    # RegEx for Price Range
    Regex5 = '<span class="priceRange.*?>(\$+)<\/span>'
    match5 = re.compile(Regex5).findall(page_content)
    
    

    for x in range(len(match5[0:10])):
        PriceRangeValue = [match5[x]]
        test[x].extend(PriceRangeValue)
    

    # RegEx for Neighborhood
    Regex6 = '<span class=" css-chan6m">([A-Z].*?)<\/span>'
    match6 = re.compile(Regex6).findall(page_content)
    
    

    for x in range(len(match6[0:10])):
        NeighborhoodValue = [match6[x]]
        test[x].extend(NeighborhoodValue)


    # RegEx for Closing Time 
    Regex7 = '<p class=" css-qgunke"><span class=" css-1l0s6bn" data-font-weight="bold">.*?<\/span> .*?([0-9]+:[0-9]+.*?)<\/p>'
    match7 = re.compile(Regex7).findall(page_content)
    
    

    for x in range(len(match7[0:10])):
        TimeValue = [match7[x]]
        test[x].extend(TimeValue)
    

    # RegEx for Reviews
    Regex8 = '<p class="css-16lklrv">(.*?)<span'
    match8 = re.compile(Regex8).findall(page_content)
    print(match8)

    for x in range(len(match8[0:10])):
        ReviewValue = [match8[x]]
        test[x].extend(ReviewValue)
        
        
# Checking length items - For reference only.  We only want complete records
    for x in test:
       print(len(x))
            
            
    # Assigning to Attribute   
    for x in test:
        if len(x) == 9:
            ID = x[0]
            REST_NAME = x[1]
            STAR_RATING = x[2]
            NUM_REVIEWS = x[3]
            CATEGORIES = x[4]
            PRICE_RANGE = x[5]
            NEIGHBORHOOD = x[6]
            CLOSING_TIME = x[7]
            REVIEW = x[8]
            query = "INSERT INTO Yelp VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?)"
            c.execute(query, (ID, REST_NAME, STAR_RATING, NUM_REVIEWS, CATEGORIES, PRICE_RANGE, NEIGHBORHOOD, CLOSING_TIME, REVIEW))
        
    # Repeat   
    i = i + 10
    print(i)
    if Page != LastPage:
        URLString = "https://www.yelp.com/search?find_desc=boston+restaurants&find_loc=Boston%2C+MA+02122&start=$NUM$"
        URL = URLString.replace("$NUM$",str(i))
        print(URL)
        print("Page:" + ' ' + str(pagenumber + 1), URL)
        driver.get(URL)  # Gets the page
        page_content = driver.page_source 
        
        # Repeat
        RegexPage = '<div class=" border-color--default__09f24__NPAKY text-align--center__09f24__fYBGO"><span class=" css-chan6m">([0-9]+) of (24)<\/span><\/div>'
        PageMatch = re.compile(RegexPage).findall(page_content)
        print(PageMatch)

        Page = int(PageMatch[0][0].strip())
        LastPage = int(PageMatch[0][1].strip())
    
    else:
        break
```

```python
# Print the individual records from the database   
c.execute("SELECT * FROM Yelp")
records = c.fetchall()
for record in records:
    print(record)
    print()
```

```python
# Commit changes, Close Cursor and Driver    
conn.commit()
conn.close()
driver.close()
```
